```python
from thundersvm import SVC
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
import os
import pickle
from collections import Counter
```

```python
with open(PATH + 'aapl_one_gram_features_labelled_df.pkl', 'rb') as f:
    labelled_aapl_news_df = pickle.load(f)
```

```python
Counter(labelled_aapl_news_df['source']).most_common()
```

```python
TOP_SOURCES_COUNT = 10
top_sources = Counter(labelled_aapl_news_df['source']).most_common()[:TOP_SOURCES_COUNT]
```

```python
# Find accuracy for each source
for source, count in top_sources:
  data = labelled_aapl_news_df.loc[labelled_aapl_news_df['source']==source]
  data.reset_index(inplace=True, drop=True)

  # Split data into training and testing set
  trainX, trainY = np.stack(data['features']), np.stack(data['label'])
  X_train, X_test, y_train, y_test = train_test_split(trainX, trainY, test_size=0.2, random_state=23)

  # Model training and prediction
  model = SVC(C=100, kernel='rbf')
  model.fit(X_train, y_train)
  # y_pred = model.predict(X_test)
  accuracy = model.score(X_test, y_test)

  print('Accuracy: {}\t Source: {}\t Dataset size: {}'.format(accuracy, source, count))
```
